Week 1 Challenge: Compare Stemming vs Lemmatization

```python
print("\nChallenge: Stem vs Lemma comparison")
test_words = ["running", "better", "studies", "feet"]

lemmatizer = WordNetLemmatizer()
stemmer = nltk.stem.PorterStemmer()

comparison = []
for word in test_words:
    lemma = lemmatizer.lemmatize(word)
    stem = stemmer.stem(word)
    comparison.append({'word': word, 'lemma': lemma, 'stem': stem})

df_comp = pd.DataFrame(comparison)
print(df_comp)
print("â†’ Lemmatization preserves meaning better for sentiment analysis")
```
