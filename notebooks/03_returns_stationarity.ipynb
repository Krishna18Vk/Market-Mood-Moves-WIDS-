```python
import yfinance as yf
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller


```

Download data (Week 1 ยง7.2)

```python

aapl = yf.download('AAPL', start='2024-01-01', end='2025-01-01')
prices = aapl['Close']
```

 Returns (Week 1 ยง3.4)

```python

returns = prices.pct_change().dropna() * 100  # %

print("AAPL Stats:")
print(f"Mean return: {returns.mean():.2f}%")
print(f"Std: {returns.std():.2f}%")

```

ADF Test ยง11.3

```python

def adf_test(series, title):
    result = adfuller(series)
    print(f"\n{title}: p-value = {result[1]:.4f}")
    print("Stationary" if result[1] < 0.05 else "Non-stationary")

adf_test(prices, "Raw Prices")
adf_test(returns, "Daily Returns")
```
